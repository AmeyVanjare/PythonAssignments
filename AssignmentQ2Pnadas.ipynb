```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

create a list for storing year 2010 to 2014
create a list for each year for storing sales amount for 5 products in each years
draw pie chart and stack graph to compare sales
1. Yearly comparison
Year and max sale in each year
2. Draw separate graph for each year(5 different pie charts)
Product and average sale of that product

```python
years = ['2010','2011','2012','2013','2014']
ProdLabel = ['Cookies','soap','Tea','boxed pasta','Bottled water']
```

```python
lst2010 = [40,50,60,10,40]
lst2011 = [15,20,25,5,5]
lst2012 = [20,20,30,10,15]
lst2013 = [15,40,10,20,6] 
lst2014 = [10,20,20,10,6]
lstYears = [lst2010,lst2011,lst2012,lst2013,lst2014]
```

1.Yearly comparison Year and max sale in each year

```python
maxList=[]
for year in lstYears:
  maxList.append(max(year))
print(maxList)
```

```python
plt.bar(years, maxList, color ='maroon',width = 0.4)
plt.show()
```

2.Draw separate graph for each year(5 different pie charts) Product and average sale of that product

```python
plt.pie(lst2010,labels=ProdLabel,autopct='%1.1f%%')
plt.title('2010')
plt.axis('equal')
plt.show()

plt.pie(lst2011,labels=ProdLabel,autopct='%1.1f%%')
plt.title('2011')
plt.axis('equal')
plt.show()

plt.pie(lst2012,labels=ProdLabel,autopct='%1.1f%%')
plt.title('2012')
plt.axis('equal')
plt.show()

plt.pie(lst2013,labels=ProdLabel,autopct='%1.1f%%')
plt.title('2013')
plt.axis('equal')
plt.show()

plt.pie(lst2014,labels=ProdLabel,autopct='%1.1f%%')
plt.title('2014')
plt.axis('equal')
plt.show()
```
